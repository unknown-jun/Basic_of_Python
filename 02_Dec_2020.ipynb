#49. 클래스 이해하기 (class)
- 객체지향 프로그램에서 중요한 단어가 바로 클래스
- 클래스는 프로그래머가 지정한 이름을 만든 하나의 독립된 공간이며, 이름 공간(name space)라고 부름
- 클래스를 구성하는 주요 요소는 클래스에서 변수 역활을 하는 클래스 맴버와 함수와 동일한 역활을 하는 클래스 메소드가 잇음

|총 설계도|총|
|--|--|
|클래스|객체|

````
# 예제1. 총 클래스( 설계도 )를 생성하시오

class Gun():
# 총의 기능: 1. 총알을 장전하는 기능, 2. 총 쏘는 기능
    def charge( self, num ): # 총알을 충전하는 함수
        self.bullet = num

    def shoot( self, num ):  # 총을 쏘는 함수
        for i in range(num):
            if self.bullet > 0:
                print('탕!')
                self.bullet -= 1
            elif self.bullet == 0:
                print('총알이 없습니다.')
                break
# 설명: 파이썬 클래스를 만들 때 self 키워드는 필수! (안쓰면 에러 발생)
# self의 뜻이 자기자신인 것처럼 이 설계도로 총을 만들고 총을 사용할 때 charge와 shoot
````

- 설명: 파이썬 클래스를 만들 때 self 키워드는 필수! (안쓰면 에러 발생)
- self는 자기자신이라는 뜻
- 이 설계도로 총을 만들고 총을 사용할 때 charge와 shoot 기능을 사용할 것인데 이때 이 총이 내꺼라는 의미가 바로 self
- 총을 생산하는 명령어
    - gun1 = Gun()    # 이준혁
    - gun2 = Gun()    # 한결
    - gun3 = Gun()    # 세원
- 내가 gun1 총을 사서 다음과 같이 충전했다.  
  
    - gun.charge(100)  

- 실제로 총 설계도에는 charge(self, num)이라고 해서 입력매개변수가 2개인데 총을 사용할 때 총전할 때는 gun1.charge(100) 이렇게 매개변수 하나에 값을 입력했음
- 그러면 self 입력매개변수엔 자동으로 gun1이 들어감

```
gun1.charge(10)

10발이 충전되었습니다
```

````
class Gun():
    def charge(self, num):
        self.bullet = num
        print(num, ' 발이 충전되었습니다.')

    def shoot(self, num):
        for i in range(num):
            if self.bullet > 0:
                print('탕!')
                self.bullet -= 1
            elif self.bullet == 0:
                print('총알이 없습니다.')
                break
gun1 = Gun()
gun1.charge(10)
````

**문제 158. 이번에는 총을 쏘면 총알이 탕!탕! 하면서 아래쪽에 '몇발 남았습니다'라는 메세지가 출력되게 하시오**

````
class Gun():
    def charge(self, num):
        self.bullet = num
        print(num, '발이 충전되었습니다.')

    def shoot(self, num):
        for i in range(num):
            if self.bullet >0:
                print('탕!')
                print(self.bullet - 1,'발 남았습니다.')
                self.bullet -= 1
            else:
                print('총알이 없습니다.')
                break
gun1 = Gun()
gun1.charge(10)
gun1.shoot(3)
````

**문제 160. 카드 클래스를 만들고 아래와 같이 카드를 충전하고 사용하시오**

```
card1 = Card()
"카드가 발급되었습니다."
card1.charge(10000)
"10000원 충전되었습니다."
card1.consume(1000)
"1000원이 사용되었습니다."
```

````
class Card():

    def __init__(self):
        self.money = 0
        print("카드가 발급되었습니다.")

    def charge(self, num):
        self.money += num
        print(num, "원 충전되었습니다.")
        
    def consume(self, num):    
        if self.money > 0:
            self.money -= num
            print(num, "원이 사용되었습니다.")
            print('잔액이', self.money, '남았습니다')
        elif self.money < 0:
            print("모든 잔액이 소진되었습니다.")
            
card1 = Card()
card1.charge(10000)
card1.consume(1000)
````

**문제 166. 위에서 만든 x에 있는 키 값들을 x축으로 두고 확률 밀도함수그래프를 생성하는데 y축의 확률 밀도함수값을 구할때 문제 164번에서 구한 평균값과 표준편차를 이용하시오**

````
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

N = 10000000
std = 5
avg = 140
list_1 = []

height = np.random.randn(N) * std + avg

for i in range(1, 10001):
    list_1.append(np.random.choice( height, 100 ).mean())

s_avg = np.mean(list_1)
s_std = np.std(list_1)
x = np.arange( 138, 142, 0.001 )
y = norm.pdf(x, s_avg, s_std)

plt.plot(x,y,color='blue')
plt.show()
````

**문제 167. 어느 도시의 초등학생의 수는전체 10만명 입니다. 학생들의 평균키는 140 이고 표준편차는 5 입니다.**  
**이 도시의 10만명의 초등학생 중 무작위로 한명을 추출했을때 이 어린이의 키가 145~ 150 (145 이상 ~ 150 미만) 사이에 있을 확률은 어떻게 될까요 ?**

````
import numpy as np

N = 100000   # 모집단의 수
std = 5      # 표준편차
avg = 140    #평균

height = np.random.randn(N) * std + avg    # 평균이 140이고 표준편차가 5인 사람들 백만명의 변수를 만듦
cnt = 0    # 확률을 구하기 위한 초기값

for i in height:        # array 자체를 for문으로 뽑아냄
    if 145 <= i < 150:  # 추출된 i 값이 145이상 150미만이면
        cnt += 1        # 총계 값에서 1을 더함

print(cnt/100000)
````
