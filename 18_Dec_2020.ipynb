#144. 파이썬에서 워드 클라우드 그리기

- 웹스크롤링한 데이터를 분석하는 방법
    1. 회사의 신제품이 출시되었을 때 소비자들의 반응을 살펴보고자 할 때( 기업 )
    2. 시기별 사회현상을 파악하고자 할 때 ( 국가 )
    3. 인공지능 상담원 (딥러닝의 RNN)을 만들기 위한 자연어 처리 학습 데이터로 웹스크롤링한 데이터를 활용
    4. 딥러닝의 CNN(인공지능의 눈)의 학습 자료로 이미지 데이터가 활용

- 데이터의 종류:
    1. 정형 데이터  : 정형화된 스키마 구조. DBMS에 저장될 수 있는 구조  
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;예) Oracle의 emp 테이블, MYSQL, MSSQL의 테이블
    2. 반정형 데이터: 데이터 내부의 데이터 구조에 대한 메타 정보가 포함된 구조의 데이터  
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;예) html, xml 문서
    3. 비정형 데이터: 웹스크롤링 기술로 수집해서 모은 데이터  
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;예) 텍스트 문서, 이미지, 동영상

**문제 433. 라라랜드 평가글 중에 평점 6점 이상이면 pos라는 비어있는 리스트에 해당 평가글을 append되게 하고 6점 미만이면 nag라는 비어있는 리스트에 평가글이 append 되게 하시오**

```
stev = open("sample6_laland_review.txt", encoding='utf8')
stev2 = stev.readlines()

pos = []
nag = []
for i in stev2:
    if int( i[6:8] ) >= 6:
        pos.append( i[8:] )   # i.strip('\n') 하면 엔터를 제거해서 결과를 보여줌
    else:
        nag.append( i[8:] )
print(pos)
```
