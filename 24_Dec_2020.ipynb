```
"""
코딩예제3: 위에서 만든 코드를 가지고 아래의 함수를 생성하시오
"""
cities = ['Jeju', 'Pangyo', 'New York', 'new york']

def cacheProcess(cities, cachesize):
    city = [i.lower() for i in cities]    # 도시 이름 다 소문자로 변경
    cache = [ None for i in range( 1, cachesize+1 ) ]  # None 4개 cache 구성
    
    for i in range(0, len(city) ):    # 0,1,2,3 
        cache.append( city[i] )
        del cache[0]
    
    return cache

print( cacheProcess(cities, 4) )
```

문제 521. 지금까지의 코드를 함수로 구현해서 아래와 같이 실행되게 하시오
~~~
print( Jaccard() )
문자열을 입력해주세요. : FRANCE
문자열을 입력해주세요. : french

16384
~~~

```
def Jaccard():
    
    str1 = input('문자열을 입력해주세요: ').upper()
    str2 = input('문자열을 입력해주세요: ').upper()

    res1 = []
    for i in range( len(str1) -1 ):
        if str1[i].isalpha() and str1[i+1].isalpha():
            res1.append( str1[i:i+2] )
  
    res2 = []
    for i in range( len(str2) -1 ):
        if str2[i].isalpha() and str2[i+1].isalpha():
            res2.append( str2[i:i+2] )

    union = set(res1 + res2)

    intersection=[]
    for i in res1:
        if i in res2:
            intersection.append(i)

    len_i = len(intersection)
    len_u = len(union)

    return int( len_i/len_u * 65536 )

print( Jaccard() )
```

문제 523. set을 이용하지 말고 아래의 A와 B 두개의 리스트의 합집합이 {1, 1, 1, 2, 2, 3, 4, 5}로 출력되게 하시오

```
A = [1, 1, 1, 2, 2, 3]
B = [1, 1, 2, 2, 4, 5]

intersection=[]
for  i  in  A:
    if  i  in B:
        intersection.append(i)
print(intersection)  

s = A + B 
diff = []
for  i  in  s:
    if  i  not  in  intersection:
        diff.append(i)

union = diff + intersection
print(union)
```
