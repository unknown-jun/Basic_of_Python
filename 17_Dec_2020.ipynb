```
# 4. 더나은 미래 첫페이지에 보이는 기사들 12개의 기사 url을 알아내기 위해 태그 이름과 클래스 이름이 무엇인지 알아내야 함 (href가 있는 곳을 찾는다)
# 태그 이름은 dic이고 클래스 이름은 elementor-post__title
# 중요!! href와 가장 가까이 있는 태그 이름과 클래스 이름을 알아내야 함
```

```

# 6. result1 리스트의 요소(BeautifulSoup으로 파싱된 html 코드들)를 for loop문으로 하나씩 빼옴
for i in result1:
    print(i)
```

```
# 9. 그러면 뽑아낸 a 태그의 html 문서의 href의 값만 얻어내시오
for i in result1:
    print( i.find_all('a')[0].get("href") )
```

```
# 10. 위의 상세기사 url을 params라는 리스트에 append 시킨다.

params = []
for i in result1:
    params.append( i.find_all('a')[0].get("href") )
```

```
# 11. 첫페이지지만 가져오는게 아니라 1페이지 부터 3페이지까지 24 X 3 = 72개의 기사 url을 params 리스트에 append 시키시오

params =[]

for i in range(1, 4):
    time.sleep(5)
    list_url = 'https://futurechosun.com/page/'+ str( i ) +'?s=%EB%B4%89%EC%82%AC'
    url = urllib.request.Request(list_url) 
    result = urllib.request.urlopen(url).read().decode("utf-8")

    soup = BeautifulSoup( result, "html.parser")

    result1=soup.find_all('div',class_ ='elementor-post__title')

    for i in result1:
        params.append( i.find_all('a')[0].get("href") )
```

```
# 더나은 미래 사이트는 사진을 클릭했을 때도 상세기사로 가고 기사제목을 클릭했을 때도 똑같이 상세기사로 가는 구조로 되어 있어서 상세 url이 중복되어서 스크롤링 때!
# 그래서 params의 요소들의 중복을 제거해야 함

my_result = set(params)   # 자료형 set은 중복값을 허용하지 않으므로 set자료형으로 만들어 중복을 제거
my_result2 = list(my_result) # 다시 리스트 형으로 변환
```

```
# 12. 위의 코드들을 가지고 bs_scroll() 이라는 함수로 생성하시오
def bs_scroll():
    params =[]

    for i in range(1, 4):
        time.sleep(5)
        list_url = 'https://futurechosun.com/page/'+ str( i ) +'?s=%EB%B4%89%EC%82%AC'
        url = urllib.request.Request(list_url) 
        result = urllib.request.urlopen(url).read().decode("utf-8")

        soup = BeautifulSoup( result, "html.parser")

        result1=soup.find_all('div',class_ ='elementor-post__title')

        for i in result1:
            params.append( i.find_all('a')[0].get("href") )
    
    return my_result2
```

```
# 8. 위의 스크립트로 eda_detail_scroll() 함수를 생성해시오
def eda_detail_scroll():
    
    list_url = 'https://www.edaily.co.kr/news/read?newsId=01098806625997864&mediaCodeNo=258'

    url = urllib.request.Request(list_url) 
    result = urllib.request.urlopen(url).read().decode("utf-8")

    soup = BeautifulSoup( result, "html.parser")

    result = soup.find_all('div', class_='news_body')

    for i in result:
        return (i.get_text())
```

```
# 10. eda_detail_scroll() 함수 안에서 eda_scroll() 함수를 호출하여 상세기사 url을 60개의 기사를 출력할 수 있도록 코드를 수정하시오
def eda_detail_scroll():
    
    list_url = eda_scroll()

    for i in list_url:
        url = urllib.request.Request(i) 
        result = urllib.request.urlopen(url).read().decode("utf-8")

        soup = BeautifulSoup( result, "html.parser")

        result = soup.find_all('div', class_='news_body')

        for i in result:
            print(i.get_text())

eda_detail_scroll()
```
